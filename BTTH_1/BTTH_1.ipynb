```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
series1 = np.random.randn(500, 1)
series2 = np.random.randn(400, 1)
print(series1)
print(series1.shape)
```

```python
# plotting first histogram
plt.hist(series1, bins = 18, label='not survived', color='blue')
 
# plotting second histogram
plt.hist(series2, bins = 40,label='survived', color='red')
plt.legend()
 
# Showing the plot using plt.show()
plt.show()
```
